# Linked List Visualization
This notebook demonstrates the visualization of Linked List operations.

```python
import sys
sys.path.append('/Users/elhadjimamadou/Desktop/Projects/data_structures')
from data_structures import LinkedList
```

```python
# Create a new linked list
ll = LinkedList()

# Insert elements
ll.insertAtBeginning(1)
ll.insertAtEnd(3)
ll.insertAfter(2, 1)

print("Initial linked list:")
ll.traverse()
ll.visualize()
```

```python
# Insert more elements
ll.insertAtEnd(4)
ll.insertAtEnd(5)

print("After adding more elements:")
ll.traverse()
ll.visualize()
```

```python
# Delete item
print("Deleting node at index 1:")
ll.deleteItem(1)
ll.traverse()
ll.visualize()
```
